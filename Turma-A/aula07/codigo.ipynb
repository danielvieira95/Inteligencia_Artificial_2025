```python
import numpy as np # importa a biblioteca numpy
from sklearn.linear_model import LinearRegression # importa o modelo de Regressao Linear
import matplotlib.pyplot as plt # importa a biblioteca matplotlib
#Dados de treinamento
vazao = np.array([10,20,30,5,35,40]).reshape(-1,1) # vazao da agua em m3/min
nivel = np.array([50,60,70,45,80,85]) # nivel do tanque em m

# Cria e treina o modelo de ML de regressão linear
modelo = LinearRegression() #  armazena o modelo de regressão linear
modelo.fit(vazao,nivel) #cria o modelo de regressao linear
```

```python
# Dados de teste
vazao_teste = np.array([15,25,33]).reshape(-1,1)
#vazao_teste = np.array([5,2,7,30,50]).reshape(-1,1)
# Fazendo predição com base na vazao de agua
nivel_previsto=modelo.predict(vazao_teste)
```

```python
# imprimir as previsoes
for i in range(len(vazao_teste)):
    print(f"Vazao:{vazao_teste[i][0]}m3/min - Nivel previsto: {nivel_previsto[i]}metros")
plt.scatter(vazao,nivel,color='blue', label='Dados de treinamento')
plt.plot(vazao_teste,nivel_previsto,color='red',label='Dados de teste')
plt.xlabel('Vazão de agua (m3/min)')
plt.ylabel('Nivel do tanque (metros)')
plt.title('Relação entre vazão de agua e nível do tanque')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# exercício 2 Regressão polinomial
import numpy as np # importa a biblioteca numpy
import matplotlib.pyplot as plt # biblioteca matplotlib
from sklearn.linear_model import LinearRegression # importa a regressao linear
from sklearn.preprocessing import PolynomialFeatures # importa a regressao polinomial

```

```python
# Dados fornecido
Q = np.array([100,200,300,400,500,600,700]).reshape(-1,1) # quantidade de calor
T = np.array([30,35,40,45,60,100,150]) # Temperatura
```

```python
print(Q)
print(T)
```

```python
# Modelo 1 Regressão Linear simples
linear_model = LinearRegression() # cria modelo linear
linear_model.fit(Q,T) # Treina o modelo
T_linear_pred = linear_model.predict(Q) # faço a previsao de temperatura com regressão linear
print(T_linear_pred)
```

```python
# Modelo 2 de Regressão Polinomial
poly = PolynomialFeatures(degree=5) # cria o modelo polinomial de grau 5
Q_poly = poly.fit_transform(Q) # treina o modelo e o _transform adequada a base de dados
poly_model = LinearRegression() # Armazena modelo
poly_model.fit(Q_poly,T)
```

```python
# Geração de valores para a curva da regressao polinomial
Q_range = np.linspace(100,700,500).reshape(-1,1) 
T_linear_curve = linear_model.predict(Q_range) # regressao linear
t_poly_curve=poly_model.predict(poly.transform(Q_range)) #Predição com regressao polinomial
# plotagem do grafico
plt.figure(figsize=(10,6))
plt.scatter(Q,T,color='black',label='Dados reais')
plt.plot(Q_range,T_linear_curve,color='green',label='Regressão Linear simples')
plt.plot(Q_range,t_poly_curve,color='blue',linestyle='--',label='Regressão Polinomial')
plt.title('Previsão de temperatura do forno com regressão')
plt.xlabel('Quantidade de calor (Q)')
plt.ylabel('Temperatura (T)')
plt.legend()
plt.grid(True)
plt.tight_layout() # ajusta layout do grafico
plt.show()
```
