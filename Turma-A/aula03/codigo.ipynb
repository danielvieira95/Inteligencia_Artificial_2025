```python
# importa as bibliotecas
import matplotlib.pyplot as plt # importa a biblioteca matplotlib
import numpy as np # importa a biblioteca numpy
```

```python
# Gerando dados para plotar o grafico
 # linspace gera valores espaçados de 10 em 10
Dados_x = np.linspace(0,10,10)
Dados_y = Dados_x * Dados_x
```

```python
plt.figure(figsize=(10,8),facecolor='gray') # define o tamanho e a cor do fundo da figura
# Cria a figura
Ax = plt.axes() # cria os eixos do grafico
Ax.set_facecolor('white') # cor do fundo do grafico
plt.plot(Dados_x,Dados_y,color='red',marker='o',linestyle='dashed',markeredgecolor='black')
plt.xlabel('Eixo x',fontsize=16) # titulo do eixo x
plt.ylabel('Eixo y',fontsize=16) # titulo do eixo y
plt.title('Gráfico simples com Matplotlib',fontsize=16) # titulo do grafico
```

```python
# Subplot 
plt.rcParams.update({'font.size':15}) # função que permite ajustar a figura
plt.figure(figsize=(12,12)) # define o tamanho para a figura
linhas =2 # 2 linhas
colunas =2 # 2 colunas
# subplot para plotar as figuras
plt.subplot(linhas,colunas,1) # plota o grafico na primeira posiçao
plt.plot(Dados_x,'r--',label ='Dados_x')
plt.legend(loc='lower right')
plt.xlabel('Eixo x')
plt.ylabel('Eixo y')
plt.title('Gráfico 1')

# Grafico 2
plt.subplot(linhas,colunas,2)
plt.plot(Dados_x,Dados_y,'b--',label='Dados x \n e y')
plt.title('Gráfico 2')
plt.xlabel('Eixo x')
plt.ylabel('Eixo y')
plt.legend()

#Grafico 3 
plt.subplot(linhas,colunas,3)
plt.plot(Dados_x*Dados_x,'g--',label='Dados_x^2')
plt.xlabel('Eixo x')
plt.ylabel('Eixo y')
plt.legend()

# Grafico 4 

plt.subplot(linhas,colunas,4)
plt.plot(Dados_y,'c--',label='Dados_y')
plt.title('Gráfico 4')
plt.xlabel('Eixo x')
plt.ylabel('Eixo y')
plt.legend()
plt.savefig('Graficos aula 03.png')
plt.show()  # remove warning
```

```python
# criando gráfico de barras

Figura = plt.figure() # cria uma variavel chamada figura
Grafico = Figura.add_axes([0,0,0.7,0.7]) # Cria eixos para o grafico
# Cria rotulos
Rotulos =['A','B','C','D','E','F','G','H','I','J']
Grafico.bar(Rotulos,Dados_x,color='red') # plota grafico de barras
Grafico.set_xlabel('Rotulos') # Adicionando titulos nos eixos
Grafico.set_ylabel('Quantidade')
```

```python
# Criando Grafico de barras azul e vermelho
Grafico.bar(Rotulos,Dados_x,color='red') # Dados 1
Grafico.bar(Rotulos,Dados_y*0.1,color='blue') # Dados 2
Figura
```

```python
Grafico.bar(Dados_x,Dados_y/2,color=['red','blue']) #exibe um grafico do lado do outro
Figura
```

```python
# Plotando boxplot
plt.boxplot(Dados_x) # Plota o boxplot
plt.grid()  # ativa a grade do grafico
plt.show()
```
