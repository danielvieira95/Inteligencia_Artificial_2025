```python
# importa as bibliotecas
import matplotlib.pyplot as plt # importa a biblioteca matplotlib
import numpy as np # importa a biblioteca numpy
```

```python
# Gerando dados para plotar o grafico
 # linspace gera valores espaçados de 10 em 10
Dados_x = np.linspace(0,10,10)
Dados_y = Dados_x * Dados_x
```

```python
plt.figure(figsize=(10,8),facecolor='gray') # define o tamanho e a cor do fundo da figura
# Cria a figura
Ax = plt.axes() # cria os eixos do grafico
Ax.set_facecolor('white') # cor do fundo do grafico
plt.plot(Dados_x,Dados_y,color='red',marker='o',linestyle='dashed',markeredgecolor='black')
plt.xlabel('Eixo x',fontsize=16) # titulo do eixo x
plt.ylabel('Eixo y',fontsize=16) # titulo do eixo y
plt.title('Gráfico simples com Matplotlib',fontsize=16) # titulo do grafico
```

```python
# Subplot 
plt.rcParams.update({'font.size':15}) # função que permite ajustar a figura
plt.figure(figsize=(12,12)) # define o tamanho para a figura
linhas =2 # 2 linhas
colunas =2 # 2 colunas
# subplot para plotar as figuras
plt.subplot(linhas,colunas,1) # plota o grafico na primeira posiçao
plt.plot(Dados_x,'r--',label ='Dados_x')
plt.legend(loc='lower right')
plt.xlabel('Eixo x')
plt.ylabel('Eixo y')
plt.title('Gráfico 1')

# Grafico 2
plt.subplot(linhas,colunas,2)
plt.plot(Dados_x,Dados_y,'b--',label='Dados x \n e y')
plt.title('Gráfico 2')
plt.xlabel('Eixo x')
plt.ylabel('Eixo y')
plt.legend()

#Grafico 3 
```
