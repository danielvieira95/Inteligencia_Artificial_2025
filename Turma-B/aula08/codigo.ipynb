```python
# importa as bibliotecas
import numpy as np #  importa a biblioteca numpy
import matplotlib.pyplot as plt # importa a matplotlib
import seaborn as sns # importa a biblioteca seaborn
from sklearn.model_selection import train_test_split # importa a biblioteca para dividir os dados em treino e teste
from sklearn.tree import DecisionTreeClassifier # importa a arvore de decisao
from sklearn import tree # importa o elemento tree para plotar a arvore
```

```python
# !pip install seaborn instalação biblioteca seaborn 
```

```python
# criando o conjunto de dados
notas = np.array([8,6,5,9,4,3,7,2,1,5,6,9,8])
classes = np.array(['bom','ruim','ruim','bom','pessimo','pessimo','bom','pessimo','pessimo',
                    'ruim','ruim','bom','bom'])
```

```python
# Dividindo a base de dados em treino e teste
notas_treino,notas_teste,classes_treino,classes_teste=train_test_split(notas.reshape(-1,1),
                                                                       classes.reshape(-1,1),
                                                                       test_size=0.2, # 20% dos dados para treino 
                                                                       random_state=42 ) # permite a reprodutibilidade dos testes
```

```python
# Criação do modelo de Machine Learning Árvore de decisão
modelo = DecisionTreeClassifier() # cria o modelo
modelo.fit(notas_treino,classes_treino) # treinamento do modelo
```

```python
# Teste do modelo
print(modelo.predict(notas_teste)) 
print(notas_teste)
```

```python
modelo.predict(np.array([4]).reshape(-1,1)) #teste de uma nota individualmente
```

```python
modelo.predict(np.array([7.5]).reshape(-1,1)) #teste de uma nota individualmente
```

```python
modelo.predict(np.array([6,5,9,7]).reshape(-1,1)) #teste de uma nota individualmente
```

```python
# Visualização do modelo de árvore de decisão
fig = plt.figure(figsize=(10,12)) # Cria uma figura do tamanho 10 x 8
# tree.plot plota a arvore de decisão
#modelo é o modelo criado do algoritmo de aprendizado supervisionado
tree.plot_tree(modelo,feature_names=notas.tolist(),class_names=classes.tolist(),filled=True)
plt.title('Árvore de decisão para classificar uma bebida \n de acordo com a nota',fontsize=18)
plt.show()
```

```python
# Plota box plot com a biblioteca seaborn
# x é o eixo que indica as classes boa, ruim ou pessima
# y indica as notas

plt.figure(figsize=(10,8))
plt.title('Box plot das notas e classes de bebidas')
sns.boxplot(x=classes, y=notas,palette=['green','yellow','red'])
plt.grid()
```

```python
# grafico de barras
plt.figure(figsize=(10,8))
sns.barplot(y=notas,x=classes,palette=['green','yellow','red'])
plt.title('Notas das bebidas',fontsize=20)
plt.xlabel('Classificação',fontsize=16)
plt.ylabel('Notas',fontsize=16)
```
