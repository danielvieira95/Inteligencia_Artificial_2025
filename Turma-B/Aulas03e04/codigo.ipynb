```python
import matplotlib.pyplot as plt # importa a biblioteca matplotlib
import numpy as np # importa a bilioteca numpy 
```

```python
# Gerando os dados para plotar o grafico
# linspace gera valores espaçados de 10 em 10
Dados_x = np.linspace(0,10,10) # gera 10 valores de 10 em 10
Dados_y = Dados_x * Dados_x
```

```python
plt.figure(figsize=(10,8),facecolor='gray') # plt.figure
#cria a figura
#figsize tamanho da figura
Ax = plt.axes() # cria os eixos do grafico
Ax.set_facecolor('white') # cor do fundo do grafico
plt.plot(Dados_x,Dados_y,color='red',marker='o',linestyle='dashed',
         markeredgecolor='black')
# plota o grafico em x e y
plt.xlabel('Eixo x',fontsize=16)
plt.ylabel('Eixo y',fontsize=16)
plt.title('Gráfico simples com Matplotlib',fontsize=16)
```

```python
# Subplot 

plt.rcParams.update({'font.size':15}) # função que permite ajustar
#figura
plt.figure(figsize=(12,12)) # define um tamanho para a figura
linhas =2 # cria 2 linhas
colunas = 2 # cria 2 colunas
# subplot para plotar as figuras
plt.subplot(linhas,colunas,1) # plota grafico na primeira posição
plt.plot(Dados_x,'r--',label='Dados_x') 
plt.legend(loc='lower right') # cria legenda para o grafico
plt.xlabel('Eixo x')
plt.ylabel('Eixo y')
plt.title('Gráfico 1')

# 2 grafico
plt.subplot(linhas,colunas,2)
plt.plot(Dados_x,Dados_y,'b--',label='Dados x \n e y')
# plota grafico alterando a cor da linha
plt.title('Grafico 2') # titulo do grafico
plt.xlabel('Eixo x')
plt.ylabel('Eixo y')
plt.legend()

# Grafico 3
plt.subplot(linhas,colunas, 3)
plt.plot(Dados_x*Dados_x,'g-',label='Dados_x^2')
plt.xlabel('Eixo x')
plt.ylabel('Eixo y')
plt.legend()

# Grafico 4
plt.subplot(linhas,colunas,4)
plt.plot(Dados_y,'c--',label='Dados_y')
plt.title('Grafico 4')
plt.xlabel('Eixo x')
plt.ylabel('Eixo y')
plt.legend()

plt.savefig('Graficos aula03.png') # serve para salvar figura
plt.show() # remove o warning


```

```python
 # criando um grafico de barras

Figura = plt.figure () # Cria uma variavel chamada figura 
Grafico = Figura.add_axes([0,0,0.7,0.7]) # cria eixos para o grafico
# cria rotulos
Rotulos=['A','B','C','D','E','F','G','H','I','J'] 
# Plota o gráfico
Grafico.bar(Rotulos,Dados_x,color='red') # plota grafico de barras
Grafico.set_xlabel('Rotulos') # adicionando titulos nos eixos
Grafico.set_ylabel('Quantidade')
```

```python
# Criando outro grafico de barras
Grafico.bar(Rotulos, Dados_x,color='red') # Dados 1
Grafico.bar(Rotulos,Dados_y*0.1,color='blue') # Dados 2
Figura
```

```python
Grafico.bar(Dados_x,Dados_y/2,color=['red','blue']) # exibe os dados um lado do outro
Figura
```
